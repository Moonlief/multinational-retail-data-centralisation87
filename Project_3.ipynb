```python
import pandas as pd
import re

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv('products.csv')

# Step 2: Extract unique values from the 'weight' column
unique_weights = df['weight'].unique()

# Step 3: Convert the unique values to a list
unique_weights_list = unique_weights.tolist()

# Print the list of unique weights
products_df = pd.DataFrame(unique_weights_list, columns=['weight'])

pd.set_option('display.max_rows', None)

print(products_df)

```

```python
products_df = products_df.dropna(subset=['weight'])

products_df['weight'] = products_df['weight'].str.replace('g .', 'g')


pd.set_option('display.max_rows', None)


print(products_df)
```

```python
def convert_to_kg(weight):


        
    def convert_single_unit(value, unit):
        if unit == 'kg':
            return float(value)
        elif unit == 'g':
            return float(value) /1000
        elif unit == 'ml':
            return float(value) * 0.001
        elif unit == 'oz':
            return float(value) * 0.0283495
        else:
            return None





    # Check for multiply pack
    if 'x' in weight:
        parts = weight.split(' x')
        
        if len(parts) == 2:
            quantity = float(parts[0].strip()) #first part
            value, unit = parts[1].strip()[:-2], parts[1].strip()[-2:]
            single_unit_weight = convert_single_unit(value, unit)
            if single_unit_weight is not None:
                return quantity * single_unit_weight
            else:
                return None
        else:
            return None
    else:
        # Handle single unit weights
        value, unit = weight[:-2], weight[-2:]
        return convert_single_unit(value, unit)

```

```python
products_df['weight_kg'] = products_df['weight'].apply(convert_to_kg)
```
